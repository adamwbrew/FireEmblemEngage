```python
characters = pd.read_csv('characters_cleaned.csv')
classes = pd.read_csv('classes_cleaned.csv')
exclusives = pd.read_csv('exclusives_cleaned.csv')
```

```python
classes_exclusive = classes.join(exclusives.set_index('Name'), on = 'Name', how = 'outer')
classes_exclusive.at[2, 'Exclusive'] = 'Fell Child'
classes_exclusive.at[19, 'Exclusive'] = 'Zephia'
classes_exclusive.fillna('Any', inplace = True)
classes_exclusive
```

```python
characters_copy = characters.set_index('Name').T
classes_exclusive_copy = classes_exclusive[list(classes_exclusive.columns[0:10]) + ['Exclusive']]
classes_exclusive_copy.set_index(['Exclusive', 'Name'], inplace = True)
classes_exclusive_copy = classes_exclusive_copy.T
```

```python
character_classes = {}
for character in characters_copy.columns:
    for character_class in classes_exclusive_copy.columns:
        if character_class[0] in [character, 'Any']:
            character_classes[(character, character_class[1])] = (characters_copy[character] + classes_exclusive_copy[character_class]).to_dict()
character_classes
```

```python
character_class_df = pd.DataFrame(character_classes)
character_class_df = character_class_df.T.reset_index()
character_class_df.rename(columns = {'level_0' : 'Name', 'level_1':'Class'}, inplace = True)
character_class_df
```

```python
character_class_df.to_csv('character_classes_test.csv', index = False)
```

```python
character_classes = []
for i in range(len(characters)):
    for j in range(len(classes_exclusive)):
        if characters.iloc[i]['Name'] == classes_exclusive.iloc[j]['Exclusive'] or classes_exclusive.iloc[j]['Exclusive'] == 'Any':
            name = pd.Series(characters.iloc[i]['Name'], index = ['Character'])
            c_class = pd.Series(classes_exclusive.iloc[j]['Name'], index = ['Class'])
            promotion = pd.Series(classes_exclusive.iloc[j]['Promotion'], index = ['Promotion'])
            stats = characters.iloc[i][1:] + classes_exclusive.iloc[j][1:-2] * (1 + (characters.iloc[i]['Name'] == 'Jean'))
            character_classes.append(pd.concat([name, c_class, stats]))

character_classes
```

```python
character_class_df = pd.DataFrame(character_classes)
character_class_df['Rating'] = character_class_df.sum(axis = 1, numeric_only = True)
character_class_df
```

```python
expected_gains = character_class_df[character_class_df.columns[:2]].merge(
    (character_class_df[character_class_df.columns[2:-1]] * 0.01 * 20).astype('int'),
    left_index = True,
    right_index = True,
    how = 'inner'
)
expected_gains
```

```python
def create_dtype_dict(input_df, output_file):
    data_dict = input_df.dtypes.to_dict()
    for key in data_dict.keys():
        if pd.api.types.is_object_dtype(data_dict[key]):
            data_dict[key] = 'varchar(50)'
        else:
            data_dict[key] = 'int'
    f = open(output_file, 'w')
    json.dump(data_dict, f)
    f.close()
```

```python
create_dtype_dict(character_classes, 'character_classes_data_dictionary.json')
create_dtype_dict(characters, 'characters_data_dictionary.json')
create_dtype_dict(expected_gains, 'expected_gains_data_dictionary.json')
```
