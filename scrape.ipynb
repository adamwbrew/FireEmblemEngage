```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
link = 'https://serenesforest.net/engage/characters/growth-rates/'
html = BeautifulSoup(requests.get(link).text, 'html.parser')
```

```python
headers = []
characters = []
for row_num, tr in enumerate(html.find('table').findAll('tr')):
    character_dict = {}
    if row_num == 0:
        for th in tr.findAll('th'):
            headers.append(th.text)
    elif tr.findAll('th') == []:
        for i, td in enumerate(tr.findAll('td')):
            if i == 0:
                character_dict[headers[i]] = td.text
            else:
                character_dict[headers[i]] = int(td.text)
        characters.append(character_dict)
```

```python
f = open('characters.csv', 'w', newline = '')

writer = csv.DictWriter(f, fieldnames = headers)
writer.writeheader()
writer.writerows(characters)
f.close()
```

```python
link = 'https://serenesforest.net/engage/classes/growth-rates/'
html = BeautifulSoup(requests.get(link).text, 'html.parser')
```

```python
headers = []
classes = []
for row_num, tr in enumerate(html.find('table').findAll('tr')):
    class_dict = {}
    if row_num == 0:
        for th in tr.findAll('th'):
            headers.append(th.text)
    elif tr.findAll('th') == []:
        for i, td in enumerate(tr.findAll('td')):
            if i == 0:
                class_dict[headers[i]] = td.text
            else:
                if td.text.strip() == '':
                    class_dict[headers[i]] = 0
                else:
                    class_dict[headers[i]] = int(td.text)
            #class_dict['Type']
        classes.append(class_dict)
```

```python
classes
```

```python
for i, c_class in enumerate(classes):
    if c_class['Name'] in ['Thief', 'Fell Child', 'Dancer']:
        c_class['Promotion'] = 'Special'
    elif i < 2 and i % 2 == 0:
        c_class['Promotion'] = 'Base'
    elif i % 2 == 1:
        c_class['Promotion'] = 'Base'
    else:
        c_class['Promotion'] = 'Advanced'

classes
```

```python
f = open('classes.csv', 'w', newline = '')

writer = csv.DictWriter(f, fieldnames = headers + ['Promotion'])
writer.writeheader()
writer.writerows(classes)
f.close()
```

```python
link = 'https://thenerdstash.com/fire-emblem-engage-list-of-all-classes/'
html = BeautifulSoup(requests.get(link).text, 'html.parser')
```

```python
exclusive_dicts = []
for i, tr in enumerate(html.findAll('table')[2].find('tbody').findAll('tr')):
    exclusive_dict = {}
    if i != 0:
        td = tr.findAll('td')
        exclusive_dict['Name'] = td[0].text
        exclusive_dict['Exclusive'] = td[1].text[:td[1].text.find(' ')]
        exclusive_dicts.append(exclusive_dict)

exclusive_dicts
```

```python
exclusive_dicts =  exclusive_dicts[:13] + [exclusive_dicts[11].copy()] + exclusive_dicts[13:]
exclusive_dicts[13]['Exclusive'] = 'Hortensia'
exclusive_dicts
```

```python
for i, exclusive_dict in enumerate(exclusive_dicts):
    if i > 0 and i % 2 == 0:
        exclusive_dict['Exclusive'] = exclusive_dicts[i-1]['Exclusive']

exclusive_dicts
```

```python
f = open('exclusives.csv', 'w', newline = '')

writer = csv.DictWriter(f, fieldnames = ['Name', 'Exclusive'])
writer.writeheader()
writer.writerows(exclusive_dicts)
f.close()
```
